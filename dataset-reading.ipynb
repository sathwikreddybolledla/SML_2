```python
import pandas as pd
import numpy as np
import os
import tensorflow as tf
import cv2
from tensorflow import keras
import random
# from tensorflow.keras import layers, Dense, Input, InputLayer, Flatten
# from tensorflow.keras.models import Sequential, Model
from  matplotlib import pyplot as plt
import matplotlib.image as mpimg
li=[]
%matplotlib inline
plt.figure(figsize=(20,20))
test_folder=r'/content/drive/MyDrive/ex dataset'
for i in range(80):
    file = random.choice(os.listdir('/content/drive/MyDrive/ex dataset'))
    image_path= os.path.join('/content/drive/MyDrive/ex dataset', file)
    img=mpimg.imread(image_path)
    li.append(img)
    ax=plt.subplot(1,80,i+1)
    ax.title.set_text(file)
    plt.imshow(img)
print(li)
```
